```python
import pandas as pd
import numpy as np
import os
import os.path as osp
import shutil
```

```python
new_data = os.listdir('./data_new/raw/Binance')
```

```python
def move_new_download_to_old_folder():
    new_data = os.listdir('./data_new/raw/Binance')
    for f in new_data:
        if f.endswith('csv'):
            folder_name = f.split('_latest')[0]
            new_dst = '../../data/data_new/raw/Binance/{}'.format(folder_name)
            if osp.exists(new_dst) == False:
                os.mkdir('../../data/data_new/raw/Binance/{}'.format(folder_name))
            file_path = './data_new/raw/Binance/{}'.format(f)
            shutil.copy(file_path,new_dst)
            
```

```python
move_new_download_to_old_folder()
```

```python
root = '../../data/data_new/raw/Binance/'
T = os.listdir(root)
for x in T:
    try:
        print(x)
        if osp.isdir(osp.join(root,x)):
            folder_name = osp.join(root,x)
            old_df = pd.read_csv(folder_name+'/{}.csv'.format(x), names = ['timestamp','open','high','low','close','volume'])
            new_df = pd.read_csv(folder_name + '/{}_latest.csv'.format(x),names=['timestamp','open','high','low','close','volume'])
            
            tmp = pd.concat([old_df,new_df])
            tmp = tmp.drop_duplicates(subset = 'timestamp')
            tmp.to_csv(folder_name + '/{}.csv'.format(x))
            
    except:
        print('{} did not work'.format(x))
```
