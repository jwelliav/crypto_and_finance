```python
import pickle

def save(l,filename):
    with open(filename, 'wb') as f:
         pickle.dump(l, f)
            
            
def load(filename):
    with open(filename, 'rb') as g:
         l = pickle.load(g)
    return l
```

```python
def get_all_data(ticker: str):
    
    start_date = 1262304000000
    end_date = 1680001367000
    
    interval = 60000*60*24*30
    
    tmp = start_date
    l = []
    
    holes = []
    
    for tmp in tqdm(range(start_date,end_date,interval)):
        try:
            aggs = client.get_aggs(
                ticker,
                1,
                "minute",
                tmp,
                tmp+interval,limit = 50000)
            l += aggs
        except:
            holes.append((tmp,tmp+interval))
        
    source_dir = './minute_data_sp_500/{}'.format(ticker)
    return l
    
    if os.path.exists('./minute_data_sp_500/{}'.format(ticker)) == False:
       os.mkdir(source_dir) 
                      
    save(l,'./minute_data_sp_500/{}/{}'.format(ticker,ticker))
    save(holes,'./minute_data_sp_500/{}/holes'.format(ticker,ticker))
    
    return l            
```
